List Comprehension - A concise syntax for creating lists that applies a function or operation to elements of an iterable

Liste Anlama - Yinelenebilir bir nesnenin öğelerine bir fonksiyon veya işlem uygulayan, listeler oluşturmak için kullanılan özlü bir sözdizimi.


Generator / Uretici - A function that returns an iterator object which yields one item at a time instead of returning a whole list - Bir listenin tamamını döndürmek yerine, her seferinde tek bir öğe üreten bir yineleyici nesnesi döndüren bir fonksiyon.

yield - A keyword used in generator functions to return a value from the function while retaining state - Üretici fonksiyonlarda, fonksiyonun durumunu koruyarak bir değer döndürmek için kullanılan bir anahtar kelime.

Iterable / Yinelenebilir nesne - An object that can return its members one at a time, allowing it to be iterated over in a loop- Üyelerini tek tek döndürebilen ve bu sayede bir döngü içinde üzerinde yineleme yapılabilen bir nesne.

Iterator / Yineleyici - An object that represents a stream of data that can be iterated over - Üzerinde yineleme yapılabilen bir veri akışını temsil eden bir nesne.

```python
# List comprehension with for loop to cube numbers
nums = [1, 2, 3, 4]
cubes = [num**3 for num in nums] 
print(cubes) # [1, 8, 27, 64]
```

```python
# Generator function yields numbers one by one 
def num_sequence(n):
    for i in range(n): 
        yield i

seq = num_sequence(5)
print(next(seq)) # 0
print(next(seq)) # 1 
```

```python
# Iterator from generator allows iteration
iterator = iter(num_sequence(3))
print(next(iterator)) # 0 
print(next(iterator)) # 1
```

```python
# Strings are iterable 
chars = ["c" for c in "hello"]
print(chars) # ['h', 'e', 'l', 'l', 'o']
```

---For dongusu icinde liste olusturma

```python
output =[]

for x in range (10):
    output.append(x**2)

print(output)
```

```python
output = [x**2 for x in range(10)]
print(output)
```

Kodların Analizi
1-  for Döngüsü (Üstteki Örnek)
2- Liste Üreteci (Alttaki Örnek)

Neden Liste Üreteci Kullanmalısın?
Sadelik: Gereksiz kod kalabalığını ortadan kaldırır.
Hız: Python bu işlemi arka planda daha optimize bir şekilde çalıştırdığı için append() metoduna göre genellikle daha hızlıdır.
Okunabilirlik: Ne yapılmak istendiği tek bir bakışta anlaşılır.

Ne Zaman for Döngüsünü Seçmelisin?
Eğer listenin içine yazacağın mantık çok karmaşıksa (iç içe 3-4 döngü veya çok fazla if-else durumu varsa), liste üreteci kullanmak kodu okunmaz hale getirebilir. Bu durumlarda geleneksel for döngüsü hata ayıklamak için daha güvenlidir.

```python
import random
scores =[]
for i in range(5):
    scores.append(random.randint(i, 10))

print(scores)
```

---Filter

```python
caps = []

for letter in "HellO WorLD":
    if letter.isupper():
        caps.append(letter)

print(caps)
```

```python
[letter for letter in "HellO WorLD" if letter.isupper()]
```

---Nest

```python
list_of_cars=[ ["Toyota", "Honda"], ["Ford", "BMW", "Audi"] , ["Mercedes", "Lexus"] ]
total_cars =[]

for sub_list in list_of_cars:
    for car in sub_list:
        total_cars.append(car)

print(total_cars)
```

```python
[car for sub_list in list_of_cars for car in sub_list]
```
