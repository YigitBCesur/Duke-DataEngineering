___Creating Numpy Arrays in Python___

ndarray

1. Size set at creation (Boyut oluşturulurken belirlenir)
Bir ndarray oluşturduğunda bellekte sabit bir yer ayrılır. Python listelerinin aksine, NumPy dizileri dinamik olarak büyüyecek şekilde tasarlanmamıştır. Boyutu değiştirmek istersen teknik olarak yeni bir dizi oluşturulur ve eskisi silinir.

2. Single data type (Tek veri tipi - Homojenlik)
Bu en kritik özelliklerden biridir. Bir ndarray içindeki tüm elemanlar aynı tipte (örneğin hepsi int64 veya float64) olmalıdır. Bu homojenlik, Python'ın her eleman için ayrı ayrı tip kontrolü yapmasını engeller ve işlemlerin vektörize edilerek çok hızlı çalışmasını sağlar.

3. Not limited to İki Boyut (Sınırsız Boyut)
Pandas DataFrame'ler genellikle 2 boyutludur (satır ve sütun). Ancak ndarray şunları yapabilir:

        1D: Bir vektör (Örn: [1, 2, 3])

        2D: Bir matris (Örn: Resimlerin siyah-beyaz piksel değerleri)

        3D ve üzeri: Tensörler (Örn: Renkli resimler - RGB kanalları veya zaman serisi verileri)

4. Can be reshaped (Yeniden Şekillendirilebilir)
Dizinin içindeki verileri ve toplam eleman sayısını değiştirmeden, görünümünü (boyutlarını) değiştirebilirsin. Örneğin, 12 elemanlı tek bir satırı, 3x4'lük bir matrise dönüştürebilirsin.

Tek Dimension'dan Cift Dimension'a cevirmek

```python
oned= np.arange(1, 21) # Create 1D array with values from 1 to 21
twod=oned.reshape(4, 5) # Reshape 1D array to 2D array with 4 rows and 5 columns
twod
```

```python
twod.reshape(2, 3) # Reshape 2D array to 2 rows and 3 columns   
# This will raise an error because the total number of elements (20) 
# cannot be reshaped into a 2x3 array (which requires 6 elements).
```

```python
twod=twod.reshape(2,2,5) # Reshape 2D array to 3D array with 2 blocks, 2 rows, and 5 columns
twod
```
