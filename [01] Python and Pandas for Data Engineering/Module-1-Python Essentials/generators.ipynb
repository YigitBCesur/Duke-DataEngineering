GENERATOR FUNCTIONS 

--return

```python
def return_numbers():
    for x in range(5):
        return x
    
return_numbers()   # sadece 0 dondurur, return fonksiyonu fonksiyonu bitirir
```

--yield

```python
def return_numbers():
    for x in range(5):
        yield x # generator fonksiyonu, yield ifadesiyle her seferinde bir deger dondurur ve kaldigi yerden devam eder  
    
return_numbers() 
```

```python
gen_num= return_numbers()
next(gen_num)
```

```python
def fibonacci():
    for current in (0, 1):
        last=current
        yield current
        while True:
             yield current
             last, current = current, last + current
```

```python
fib = fibonacci()
fib
```

fonksiyonu çağırdığında neden hemen bir sayı almadığını açıklıyor:

<generator object fibonacci at 0x7f72a43a9970>

Anlamı: Kodun içinde yield anahtar kelimesi olduğu için Python bu fonksiyonu normal bir fonksiyon gibi değil, bir üreteç olarak tanımlar.

fib = fibonacci() satırını çalıştırdığında, fonksiyonun içindeki kodlar henüz çalışmaz. Sadece o kodları sırayla çalıştırabilecek bir "kumanda" (nesne) oluşturulur.

Gördüğün o uzun sayı (0x7f7...), bu nesnenin bilgisayarın belleğindeki adresidir.

```python
next(fib)
```

kumandanın "oynat" tuşuna bastığında ne olduğunu gösteriyor:

0

Anlamı: next(fib) komutunu kullandığında, Python fonksiyonun içine girer ve ilk yield ifadesini görene kadar kodu çalıştırır.

Kodundaki for current in (0, 1): döngüsü başlar, current değeri 0 olur ve yield current satırına gelindiğinde durup dışarıya 0 değerini fırlatır.

Önemli Fark: Normal fonksiyonlar return ile işini bitirir ve ölür. Üreteçler ise yield ile kaldığı yeri hatırlar, bir sonraki next() çağrısını bekler.

List Comprehension Mantığı: 

Bu yapıyı şu şekilde formülleştirebilirsin:

[ ne yapılacak for eleman in kaynak if koşul ]

```python
#Write a generator function which will return the string 'foo' every fifth call, and 'moo' otherwise.

def foo_moo():
    i = 0
    while True:
        if i % 5 == 0:
            yield 'foo'
        else:
            yield 'moo'
        
        # yield işleminden sonra sayaç bir artar ve döngü başa döner
        i += 1

# Üreteç nesnesini oluşturuyoruz
fm = foo_moo()

# 11 kez çağırarak sonucu test ediyoruz
for _ in range(11):
    print(next(fm))
```

```python
#A comprehension using curly brackets will produce a dictionary. Set the variable 'name' to your name and use it to create a dictionary

# 1. Değişkeni kendi adınla tanımla
name = 'yigit'

# 2. Harfleri ve sayıları (indeksleri) birleştir
# list(name) -> ['y', 'i', 'g', 'i', 't']
# range(len(name)) -> [0, 1, 2, 3, 4]
items = zip(list(name), list(range(len(name))))

# 3. Süslü parantez kullanarak sözlüğü oluştur
# {anahtar: değer for döngüsü}
my_dict = {x: y for x, y in items}

# Sonucu yazdır
print(my_dict)
```

Üreticiler, `yield` ifadesiyle yineleyicileri kolayca oluşturmaya olanak tanır.

Jeneratörler, bir listenin tamamını oluşturmak yerine, öğeleri tek tek tembelce üretirler.

Yinelenebilir nesneler ve yineleyiciler önemli temel kavramlardır.

-Üreticiden çıkan ilk 5 öğeyi, tüm diziyi kaydetmeden yazdırın.

-Listedeki tek sayıları filtrelemek için liste anlama ifadesi yazın.

```python
# Generator function demonstrating yield

def num_sequence(n):
    """
    Generate sequence of numbers 
    up to n yielding one at a time
    """
    i = 0
    while i < n:
        yield i 
        i += 1

# Test generator function        
seq = num_sequence(5) 

print(next(seq)) # Print next number 
print(list(seq)) # Materialize remaining sequence
```
