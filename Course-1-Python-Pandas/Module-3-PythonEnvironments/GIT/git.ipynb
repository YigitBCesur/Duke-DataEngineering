Sürüm kontrolü  - Kaynak kodda zaman içinde yapılan değişikliklerin izlenmesi ve yönetilmesi. Birden fazla kişi arasında iş koordinasyonuna olanak tanır.

Depo (repo)  - Sürüm kontrolü tarafından izlenen bir proje klasörü. Proje dosyalarını ve revizyon geçmişini içerir.

Commit  - Bir projede yapılan bireysel bir değişiklik ve bu değişikliği açıklayan bir commit mesajı. Commit'ler, deponun belirli bir zamandaki anlık görüntüsünü yakalar.

Branch  - İzole geliştirme için bir deponun paralel sürümü. Ana kodu etkilemeden denemeler yapmak için kullanışlıdır.

Birleştirme (Merge) - Ayrı dallardaki değişiklikleri birleştirme. Farklılaşan dalları tekrar bir araya getirmek için kullanılır.

Çekme isteği (Pull request) - Bir daldaki kodu başka bir dala birleştirme isteği. Birleştirmeden önce tartışma ve incelemeye olanak tanır.

```python
# Clone an existing repo from GitHub 
git clone https://github.com/user/repo.git

# Create a new local repo
git init my-new-repo

# Check status of changes in repo
git status 

# Stage files to commit
git add file1.txt file2.txt

# Commit changes with message
git commit -m "Updated feature X"

# Create a new branch
git branch new-feature

# Switch to another branch 
git checkout another-branch

# Merge new-feature branch into main
git checkout main
git merge new-feature
```
