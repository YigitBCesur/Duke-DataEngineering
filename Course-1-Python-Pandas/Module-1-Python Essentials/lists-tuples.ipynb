Adding and Removing

```python
composers=['Vivaldi','Mozart']
composers
```

```python
composers.append('Brahms')
composers
```

```python
composers.insert(0,'Bach') # O dizinine bACH EKLEMEK İSTİYORUZ
composers
```

```python
composers.pop(0) # 0. indexteki elemanı silmek istiyoruz
composers
```

```python
vips=['Brad Pitt','Angelina Jolie','Leonardo DiCaprio']
vips
```

```python
vips.extend(composers) # vips listesine composers listesindeki elemanları eklemek istiyoruz
vips
```

```python
vips.extend(['Johnny Depp','Scarlett Johansson'])
vips
```

```python
lists=[[]]*2
lists
```

```python
lists[-1].append(4) # 2. indexteki listeye 4 eklemek istiyoruz
lists
```

```python
lists=[[] for _ in range(4)] # 4 tane boş liste oluşturmak istiyoruz, her bir liste birbirinden bağımsız olsun istiyoruz    
lists[-1]  .append(4) # 4 . indexteki listeye 4 eklemek istiyoruz, diğer liste etkilenmesin istiyoruz
lists
```
