```python
import sys
r = range(5)

sys.getsizeof(r) #sys.getsizeof(r) fonksiyonu, objenin bellekteki gerçek boyutunu byte cinsinden verir.
```

```python
r= range(1000000)
sys.getsizeof(r)
```

```python
L=list(range(1000000))
sys.getsizeof(L)
```

İşte beklediğimiz sonuç tam olarak buydu! Paylaştığın bu yeni ekran görüntüsü, range ile list (liste) arasındaki devasa farkı çok net bir şekilde ortaya koyuyor.

Bellek Analizi
range(1000000): Önceki ekran görüntüsünde gördüğümüz gibi sadece 48 byte kaplıyordu.

list(range(1000000)): Burada ise sonuç 8.000.056 byte (yani yaklaşık 8 MB).

Neden Bu Kadar Büyük Bir Fark Var?
range (Tembel/Lazy): Sadece başlangıç, bitiş ve adım sayısını saklar. Bir "tarif" gibidir; yemeğin kendisi değil, nasıl yapılacağının bilgisidir.

list (Gerçek/Eager): list(range(...)) komutunu verdiğinde, Python'a "Hadi bu 1 milyon sayının her birini gerçekten oluştur ve belleğe (RAM) yan yana diz" demiş olursun. Python her bir sayı için bellekte yer ayırır ve bu da boyutu anında 166.000 katına çıkarır.

Veri Bilimi ve MLOps Açısından Önemi
Büyük veri setleriyle (Big Data) çalışırken bu fark hayati önem taşır:

Milyarlarca satırlık bir döngüyü range ile saniyeler içinde ve sıfır bellek yüküyle çalıştırabilirsin.

Aynı şeyi bir listeye çevirmeye kalkarsan, sisteminin belleği (RAM) yetmeyebilir ve programın çöker (MemoryError).
