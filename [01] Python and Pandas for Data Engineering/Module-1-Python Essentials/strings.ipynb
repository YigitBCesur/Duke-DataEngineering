Content Type

```python
"abc123".isalpha() # isalpha metodu stringin sadece harflerden oluşup oluşmadığını kontrol eder. Stringde rakam olduğu için False döndürür.
```

```python
"abc123".isalnum() # isalnum metodu stringin sadece harf ve rakamdan oluşup oluşmadığını kontrol eder. Stringde rakam olduğu için True döndürür.
```

```python
"abc123".islower()
```

```python
"abc123".isupper()
```

```python
one="1"

one.isdigit() # isdigit metodu stringin sadece rakamlardan oluşup oluşmadığını kontrol eder. Stringde sadece rakam olduğu için True döndürür.
```

```python
one.isnumeric() # isnumeric metodu stringin sadece rakamlardan oluşup oluşmadığını kontrol eder. Stringde sadece rakam olduğu için True döndürür. isdigit ve isnumeric metotları birbirine çok benzer. İkisi de stringin sadece rakamlardan oluşup oluşmadığını kontrol eder. Ancak isnumeric metodu daha geniş bir karakter kümesini kabul eder. Örneğin, isdigit metodu sadece 0-9 arasındaki rakamları kabul ederken, isnumeric metodu ayrıca diğer dillerdeki rakamları da kabul eder. Bu nedenle, isnumeric metodu isdigit metoduna göre daha esnek bir şekilde rakamları tanır.
```

Splitting and Joining

```python
words_string =" here , are , some , words"
words_string
```

```python
words= words_string.split(",") #split metodu stringi belirtilen karaktere göre böler ve bir liste döndürür. Bu örnekte, string virgül karakterine göre bölünür ve her bir kelime bir liste elemanı olarak döndürülür.
words
```

```python
":".join(words) #join metodu bir listeyi belirtilen karakterle birleştirir ve bir string döndürür. Bu örnekte, words listesindeki elemanlar ":" karakteriyle birleştirilir ve bir string olarak döndürülür.
```

Split on new line

```python
multiline= "Bazen bazi dokumanlarda\n birden fazla satir olabilir.\n Bu durumda\n multiline string kullanmak\n isabetli olur."
multiline
```

```python
multiline.splitlines() #splitlines metodu stringi satır sonu karakterlerine göre böler ve bir liste döndürür. Bu örnekte, stringdeki her bir satır bir liste elemanı olarak döndürülür.
```
