```python
import pandas as pd
import matplotlib.pyplot as plt

# Create dummy DataFrame
data = {'name': ['John', 'Mary', 'Sam', 'Sarah'], 
        'height': [183, 167, 174, 160],
        'weight': [89, 56, 68, 50]}
df = pd.DataFrame(data)

# Scatter plot height vs weight
df.plot.scatter(x='height', y='weight')

# Set labels and title
plt.xlabel('Height (cm)') 
plt.ylabel('Weight (kg)')
plt.title('Height vs Weight')

plt.tight_layout()
plt.show()

# Function to filter by height range 
def filter_height(df, min_ht, max_ht):
    return df[(df['height'] >= min_ht) & (df['height'] <= max_ht)]

# Filter rows between 160cm and 170cm height
result = filter_height(df, 160, 170)
print(result)
```


___ACCESSING DATA IN DATAFRAME____

```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
```

```python
# Comparison operators 
print(df[df['A'] == 2]) # Simple filter
```

```python
# Boolean operators
filter = (df['A'] > 1) & (df['B'] == 6)  
print(df[filter]) # Complex filter
```

```python
# Filters with loc accessor
filter = df['A'] > 1
print(df.loc[filter]) 
```

isin() method

```python
# isin() method
values = [2, 5]
filter = df['A'].isin(values) 
print(df[filter])
```

___UPDATING DATAFRAME___

```python
data= {'First': ['Alice', 'Bob', 'Charlie'], ''
'Last': ['Smith', 'Johnson', 'Williams'],
'city': ['New York', 'Los Angeles', 'Chicago'],
'total_sales_in_working_years': [101200, 159000, 212000], 
'working_years': [5, 3, 8]}

clients = pd.DataFrame(data)
clients
```

adding rows  --append()

```python
data1= {'First': ['Kallie', 'Mary'], 
        'Last': ['Jim', 'Jonny'],
         'city': ['Dallas', 'NewOrleans'],
'total_sales_in_working_years': [121900, 215670], 
'working_years': [4, 6]}

new_clients= pd.DataFrame(data1)
new_clients
```

```python
### HATALI### clients.append(new_clients) # artik 'append' yerine 'concat' kullanıyoruz

clients = pd.concat([clients, new_clients], ignore_index=True)

clients.tail()
```

Setting specific value - Bir degeri degistirmek

```python
clients.loc[1, 'First']='Frankie'  #loc ile belirli bir hücreye erişip değerini değiştirebiliriz. loc[row_index, column_name] şeklinde kullanılır. Bu örnekte, 1. satırın 'First' sütunundaki değeri 'Frankie' olarak değiştirdik.
clients   #Yani BOB , FRANKI oldu
```

```python
clients.loc[0:1 ,'working_years' ]= 7 #loc ile belirli bir hücreye erişip değerini değiştirebiliriz. loc[row_index, column_name] şeklinde kullanılır. Bu örnekte, 0 ve 1. satırların 'working_years' sütunundaki değeri 7 olarak değiştirdik.
clients    #0 ve 1. satırların 'working_years' sütunundaki değeri 7 oldu.
```

Math Operations

```python
average_sales=clients.total_sales_in_working_years / clients.working_years
average_sales

clients['average_sales'] = average_sales
clients
```

Replace

```python
clients.replace('Alice', 'Alicia', inplace=True)  #replace() methodu ile belirli bir değeri başka bir değerle değiştirebiliriz. inplace=True ile değişiklikleri kalıcı hale getiririz.
clients.replace({'city': 'New York'}, {'city': 'NYC'}, inplace=True)  #city sütununda 'New York' değerini 'NYC' olarak değiştirdik.
clients
```
