Dictionary -Sözlük - Veri değerlerini depolamak için kullanılan, sıralanmamış anahtar-değer çiftleri koleksiyonu. Sözlükler süslü parantezler {} ile yazılır ve iki nokta üst üste ile ayrılmış anahtarlar ve değerler içerir. 

Key - Anahtar - Sözlükteki değerlere erişmek için kullanılan benzersiz tanımlayıcı. Anahtarlar dizeler, sayılar veya diğer değiştirilemez nesneler olabilir. 

Value - Değer - Her anahtarla ilişkili veri. Değerler sayılar, metin dizeleri, listeler veya diğer nesneler olabilir. 

```python
# Create a dictionary
dict1 = {"key1": "value1", "key2": 2} 

# Access value by key  
print(dict1["key1"])

# Add new key-value
dict1["key3"] = "value3"
print(dict1)
```

Set -Küme  - Benzersiz öğelerden oluşan, sıralanmamış bir koleksiyon. Kümeler süslü parantezler {} ile yazılır.

Set operations - Küme işlemleri  - Birleşim, kesişim, fark ve simetrik fark gibi kümelerle etkileşim kurma yöntemleri.

```python
# Create a set
set1 = {"value1", "value2", "value3"}

# Add new element  
set1.add("value4")
print(set1)

# Set union
set2 = {"value4", "value5", "value6"}
set3 = set1 | set2 
print(set3) 

# Set intersection 
set4 = set1 & set2
print(set4)

# Set difference
set5 = set1 - set2 
print(set5)
```
